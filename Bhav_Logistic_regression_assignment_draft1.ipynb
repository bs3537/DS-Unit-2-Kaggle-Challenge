```python
#Apply one hot encoding on train, validation and test datasets
import category_encoders as ce

encoder = ce.OneHotEncoder(use_cat_names=True)
hot_train = encoder.fit_transform(train)
hot_val = encoder.fit_transform(val)
hot_test = encoder.transform(test)
```

Begin with baselines for classification

```python
# Train dataset baseline
target = 'Great'
y_train = hot_train[target]
y_train.value_counts(normalize=True)
```

```python
#In train dataset, 40.9% burritos are Great.
```

```python
majority_class = y_train.mode()[0]
y_pred = [majority_class] * len(y_train)
y_pred
```

```python
#baseline accuracy if we guessed the majority class for every prediction on train dataset

from sklearn.metrics import accuracy_score
accuracy_score(y_train, y_pred)
#The output is 59.06%.
```

```python
#baseline accuracy if we guessed the majority class for every prediction on validation dataset.
y_val = hot_val[target]
y_pred = [majority_class] * len(y_val)
accuracy_score(y_val, y_pred)
#The output is 55.29%.
```

Use scikit-learn for logistic regression.

```python
# 1. Import estimator class
from sklearn.linear_model import LogisticRegression
```

```python
# 2. Instantiate this class
log_reg = LogisticRegression(random_state=42)
```

```python
# 3. Arrange X feature matrices (already did y target vectors)
features = ['Tortilla', 'Temp', 'Meat', 'Fillings', 'Salsa', 'Synergy', 'Wrap', 'Cost', 'Meat', 'Meat:filling']
X_train = hot_train[features]
X_val = hot_val[features]
```

```python
# 4. Scale the data before doing logistic regression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# 5. Impute missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
X_train_imputed = imputer.fit_transform(X_train_scaled)
X_val_imputed = imputer.transform(X_val_scaled)
```

```python
# 4. Fit the model
log_reg.fit(X_train_imputed, y_train)
```

```python

# 5. Apply the model to new data.
# The predictions look like this ...
print('Validation Accuracy', log_reg.score(X_val_imputed, y_val))
```

```python
#So, after applying logistic regression using the above features, our validation accuracy improved to 83.5%.
```

```python
# The predictions look like this
log_reg.predict(X_val_imputed)
```
